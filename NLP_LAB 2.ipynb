```python
# WORD TOKENIZE
texts = """Snowball is a small string processing programming language designed for creating stemming algorithms for use in information retrieval.
The basic datatypes handled by Snowball are strings of characters, signed integers, and boolean truth values, or more simply strings, integers and booleans."""

for text in texts:
    sentences = nltk.sent_tokenize(texts)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        print(words)
        tagged = nltk.pos_tag(words)
        print(tagged)
```
