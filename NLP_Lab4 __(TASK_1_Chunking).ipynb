```python
#Chunking

import nltk
```

```python
text= open("E:/3rd_sem/NLP/lab/LoR Sample 1.txt")
data = text.read()

sentences = nltk.sent_tokenize(data)
grammer = "NP: {<DT>?<JJ>*<NN>}"

chunk= nltk.RegexpParser(grammer)

for i in sentences:
    words =  nltk.word_tokenize(i)
    tagged = nltk.pos_tag(words)
    chunked = chunk.parse(tagged)
    print(chunked)
```

```python
chunked.draw()
```

```python
# NER

import spacy
nlp = spacy.load('en_core_web_sm')


doc = nlp(data)


from spacy import displacy

displacy.render(doc, style='ent')
```
