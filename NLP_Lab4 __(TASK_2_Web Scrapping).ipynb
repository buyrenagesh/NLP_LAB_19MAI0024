# Basic Learnings in web scarapping

```python
import urllib.request
```

```python
wiki = "https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India"
```

```python
page = urllib.request.urlopen(wiki)
```

```python
soup = BeautifulSoup(page)
soup
```

```python
# Extracting the tag of the web article

soup.title
```

```python
#Extracting the string of the title
title= soup.title.string
title
```

```python
#Find all the links within page’s <a> tags

soup.a
```

```python
#List of all links

soup.find_all("a")
```

```python
#Access all tables
all_tables=soup.find_all('table')
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable plainrowheaders')
right_table
```

```python
#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    states=row.findAll('th') #To store second column data
    if len(cells)==6: #Only extract table body not heading
        A.append(cells[0].find(text=True))
        B.append(states[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
        G.append(cells[5].find(text=True))
```

```python
#import pandas to convert list to data frame
import pandas as pd
df=pd.DataFrame(A,columns=['Number'])
df['State/UT']=B
df['Admin_Capital']=C
df['Legislative_Capital']=D
df['Judiciary_Capital']=E
df['Year_Capital']=F
df['Former_Capital']=G
df.head()
```

```python
import numpy as np
#To replace all line breaks in all textual columns
for col in df.columns: 
    if df[col].dtype == np.object_:
        df[col] = df[col].str.replace('\n','');
```

# With Paragraph

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.bbc.com/sport/football/46897172"
```

```python
# We use try-except incase the request was unsuccessful because of 
# wrong URL
try:
   page = urlopen(url)
except:
   print("Error opening the URL")
```

```python
#HTML Parsing
soup = BeautifulSoup(page, 'html.parser')
```

```python
#Find required Content
content = soup.find('div', {"class": "story-body sp-story-body gel-body-copy"})
```

```python
# Store in string the content with tag <p> (paragraph)
article = ''
for i in content.findAll("p"):
    article = article + ' ' +  i.text
```

```python

with open('scraped_text.txt', 'w') as file:
    file.write(article)
```

```python
text= open("scraped_text.txt")
data = text.read()
sentences = nltk.sent_tokenize(data)
grammer = "NP: {<DT>?<JJ>*<NN>}"

chunk= nltk.RegexpParser(grammer)

for i in sentences:
    words =  nltk.word_tokenize(i)
    tagged = nltk.pos_tag(words)
    chunked = chunk.parse(tagged)
    print(chunked)
```

```python
data
```

```python
chunked.draw()
```
